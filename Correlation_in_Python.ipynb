# Correlation in Python

## Import necessary libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import seaborn as sns
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)
````

## Read Data

````python
df = pd.read_csv('movies.csv')
````

````python
df.head()
````

## Explore Data Analysis

### Missing data

````python
for col in df.columns:
    pct_missing  = np.mean(df[col].isnull())
    print("{} - {}%".format(col, pct_missing))
````

### Drop NA values

````python
df = df.dropna()
````

### Categorical and numerical data type

````python
# All data types
df.dtypes
````

#### Categorical data type

````python
df.select_dtypes(include='object').columns
````

````python
df.select_dtypes(exclude='object').columns
````

## Data Cleaning

### Convert ```budget, gross``` columns to interger data type

````python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
````

### Correct the year released

````python
df['year_correct'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
````

````python
df.head()
````

````python
df.sort_values(by=['gross'], inplace=False, ascending=False)
````

````python
pd.set_option('display.max_rows', None)
````

## EDA (Cont)

### Correlation between Budget and Company

````python
# Scatter plot with budget and gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget and Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
````

### Plot the budget vs gross using seaborn

````python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":'red'}, line_kws={"color":'blue'});
````

### Correlation of numeric features

````python
df.corr(method='kendall') # pearson, kendall, spearman
````

````python
# -> Hign correlation between budget and gross
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movies features')
plt.ylabel('Movies features')
plt.show()
````

### Correlation of all of features together

````python
df_numerized = df
for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype=='object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
````

````python
corr_mat_numerized = df_numerized.corr()
corr_mat_numerized
````

````python
corr_pairs = corr_mat_numerized.unstack()
corr_pairs
````

````python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
````

````python
high_corr = sorted_pairs[(sorted_pairs > 0.5)]
high_corr
````
